```python
import nltk
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
input="hi i am Ananya. How r u"
output1=word_tokenize(input)
print(output1)
```

```python
input="hi hello bye !"
output2=sent_tokenize(input)
print(output2)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
input = "this is the second session"
doc = nlp(input)
```

```python
for token in doc:
  print(token)
```

```python
for sent in doc.sents:
  print(sent)
```
